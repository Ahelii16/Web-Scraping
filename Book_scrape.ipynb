```python
from urllib.request import urlopen
from bs4 import BeautifulSoup as bs
```

```python
column_titles = ["image_url", "book_title", "product_price"]

with open("books.csv", 'w', encoding = 'utf-8') as f:
    header_str = ','.join(column_titles)
    header_str += '\n'
    f.write(header_str)

    for i in range(1, 51):
        url = "http://books.toscrape.com/catalogue/page-" + str(i) + ".html"
        data = urlopen(url).read()
        soup_obj = bs(data, 'html.parser')
        books = soup_obj.findAll("li", {"col-xs-6 col-sm-4 col-md-3 col-lg-3"})
        
        for b in books:
            row_str = ""
            productDivs = b.find('div', attrs={'class' : 'image_container'})
            row_str = productDivs.find('img')['src'] + ","
            title = productDivs.find('img')['alt']
            title = title.replace(',', '')
            print(title)
            row_str = row_str + title + ","
            
            productDivs = b.find('p', attrs={'class' : 'price_color'}).text
            row_str = row_str + productDivs
            
            row_str += '\n'
            f.write(row_str)
```
